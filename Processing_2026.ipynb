```python
'''Hàm xóa khoảng trắng thừa'''
def remove_space(text):
    if isinstance(text, str):
        '''Loại bỏ khoảng trắng thừa và các kí tự đặc biệt ở đầu và cuối'''
        text = text.strip(" ,.x%\xa0")
        '''Loại bỏ khoảng trắng thừa ở giữa'''
        text = re.sub(r'\s+', ' ', text)
        '''Loại bỏ dấu -, nhưng phải tách riêng ra câu lệnh khác vì để chung strip sẽ bị mất số âm'''
        if text == '-':
            text = text.strip("-")
        return text
    return text
```

```python
def remove_dot_between_digits(x):
    return re.sub(r'(?<=\d)\.(?=\d)', '', str(x))
```

```python
def convert_datetime(date, f = "%d/%m/%Y"):
    try:
        date = datetime.strptime(date, f)
        return date.strftime("%Y-%m-%d %H:%M:%S")
    except ValueError:
        return date
```

```python
def convert_datetime_r3(date, f = "%d/%m/%Y"):
    try:
        date = datetime.strptime(date, f)
        return date.strftime("%Y-%m-%d %H:%M:%S")
    except:
        pass
    try:
        dt = datetime.strptime(date, "%Y-%d-%m %H:%M:%S")
        return dt.strftime("%Y-%m-%d %H:%M:%S")
    except:
        pass
```

```python
def processing(file_name, output):
    if file_name == HMGPXL or file_name == TMTDM or file_name == HMV:
        df_ref = pd.read_excel(file_name, sheet_name = '1.1 Referral_Daily Input', dtype = str)
        df_partner = pd.read_excel(file_name, sheet_name = '1.2 Referral_Partner List',  dtype = str)
        df_chc = pd.read_excel(file_name, sheet_name = '2. CHC_Daily Input', dtype = str, skiprows=[1])
        df_b2c = pd.read_excel(file_name, sheet_name = '3. B2C_Daily Input', dtype = str)
    elif file_name == HPEC:
        df_ref = pd.read_excel(file_name, sheet_name = '1.1 Referral_Daily Input', dtype = str, skiprows=[1])
        df_partner = pd.read_excel(file_name, sheet_name = '1.2 Referral_Partner List',  dtype = str, skiprows=[1])
        df_chc = pd.read_excel(file_name, sheet_name = '2. CHC_Daily Input', dtype = str, skiprows=[1])
        df_b2c = pd.read_excel(file_name, sheet_name = '3. B2C_Daily Input', dtype = str)
    elif file_name == HPH:
        df_ref = pd.read_excel(file_name, sheet_name = '1.1 Referral_Daily Input', dtype = str)
        df_partner = pd.read_excel(file_name, sheet_name = '1.2 Referral_Partner List',  dtype = str, skiprows=[1])
        df_chc = pd.read_excel(file_name, sheet_name = '2. CHC_Daily Input', dtype = str, skiprows=[1])
        df_b2c = pd.read_excel(file_name, sheet_name = '3. B2C_Daily Input', dtype = str)
    elif file_name == HMBD:
        df_ref = pd.read_excel(file_name, sheet_name = '1.1 Referral_Daily Input', dtype = str, skiprows=[1])
        df_partner = pd.read_excel(file_name, sheet_name = '1.2 Referral_Partner List',  dtype = str)
        df_chc = pd.read_excel(file_name, sheet_name = '2. CHC_Daily Input', dtype = str)
        df_b2c = pd.read_excel(file_name, sheet_name = '3. B2C_Daily Input', dtype = str, skiprows=[1,2])
    elif file_name == HMBP:
        df_ref = pd.read_excel(file_name, sheet_name = '1.1 Referral_Daily Input', dtype = str, skiprows=[1])
        df_partner = pd.read_excel(file_name, sheet_name = '1.2 Referral_Partner List',  dtype = str)
        df_chc = pd.read_excel(file_name, sheet_name = '2. CHC_Daily Input', dtype = str, skiprows=[0])
        df_b2c = pd.read_excel(file_name, sheet_name = '3. B2C_Daily Input', dtype = str, skiprows=[0])
    elif file_name == HMCL:
        df_ref = pd.read_excel(file_name, sheet_name = '1.1 Referral_Daily Input', dtype = str, skiprows=[1])
        df_partner = pd.read_excel(file_name, sheet_name = '1.2 Referral_Partner List',  dtype = str)
        df_chc = pd.read_excel(file_name, sheet_name = '2. CHC_Daily Input', dtype = str)
        df_b2c = pd.read_excel(file_name, sheet_name = '3. B2C_Daily Input', dtype = str, skiprows=[0])
    elif file_name == HMDNI:
        df_ref = pd.read_excel(file_name, sheet_name = '1.1 Referral_Daily Input', dtype = str, skiprows=[0])
        df_partner = pd.read_excel(file_name, sheet_name = '1.2 Referral_Partner List',  dtype = str)
        df_chc = pd.read_excel(file_name, sheet_name = '2. CHC_Daily Input', dtype = str, skiprows=[1])
        df_b2c = pd.read_excel(file_name, sheet_name = '3. B2C_Daily Input', dtype = str, skiprows=[1,2])
    else:
        df_ref = pd.read_excel(file_name, sheet_name = '1.1 Referral_Daily Input', dtype = str)
        df_partner = pd.read_excel(file_name, sheet_name = '1.2 Referral_Partner List',  dtype = str)
        df_chc = pd.read_excel(file_name, sheet_name = '2. CHC_Daily Input', dtype = str)
        df_b2c = pd.read_excel(file_name, sheet_name = '3. B2C_Daily Input', dtype = str)
    for col in df_ref.columns:
        if df_ref[col].dtype != 'datetime64[ns]':# and df_ref[col].dtype != 'float64':
            df_ref[col] = df_ref[col].astype(str)
            df_ref[col] = df_ref[col].apply(remove_space)
            df_ref[col] = df_ref[col].str.upper()
            df_ref[col] = df_ref[col].replace('NAN','')
    for col in df_partner.columns:
        if df_partner[col].dtype != 'datetime64[ns]':# and df_partner[col].dtype != 'float64':
            df_partner[col] = df_partner[col].astype(str)
            df_partner[col] = df_partner[col].apply(remove_space)
            df_partner[col] = df_partner[col].str.upper()
            df_partner[col] = df_partner[col].replace('NAN','')
    for col in df_chc.columns:
        if df_chc[col].dtype != 'datetime64[ns]':# and df_chc[col].dtype != 'float64':
            df_chc[col] = df_chc[col].astype(str)
            df_chc[col] = df_chc[col].apply(remove_space)
            df_chc[col] = df_chc[col].str.upper()
            df_chc[col] = df_chc[col].replace('NAN','')
    for col in df_b2c.columns:
        if df_b2c[col].dtype != 'datetime64[ns]':# and df_b2c[col].dtype != 'float64':
            df_b2c[col] = df_b2c[col].astype(str)
            df_b2c[col] = df_b2c[col].apply(remove_space)
            df_b2c[col] = df_b2c[col].str.upper()
            df_b2c[col] = df_b2c[col].replace('NAN','')
    
    df_ref[df_ref.columns[10]] = df_ref[df_ref.columns[10]].astype(str).apply(convert_datetime).astype('datetime64[ns]')
    df_ref[df_ref.columns[11]] = df_ref[df_ref.columns[11]].astype(str).apply(convert_datetime).astype('datetime64[ns]')
    df_ref[df_ref.columns[12]] = df_ref[df_ref.columns[12]].astype(str).apply(convert_datetime).astype('datetime64[ns]')

    df_partner[df_partner.columns[8]] = df_partner[df_partner.columns[8]].astype(str).apply(convert_datetime).astype('datetime64[ns]')

    df_chc[df_chc.columns[9]] = df_chc[df_chc.columns[9]].astype(str).apply(convert_datetime).astype('datetime64[ns]')
    df_chc[df_chc.columns[14]] = df_chc[df_chc.columns[14]].astype(str).apply(convert_datetime).astype('datetime64[ns]')
    df_chc[df_chc.columns[20]] = df_chc[df_chc.columns[20]].astype(str).apply(convert_datetime).astype('datetime64[ns]')
    df_chc[df_chc.columns[23]]  = df_chc[df_chc.columns[23]].astype(str).apply(convert_datetime).astype('datetime64[ns]')

    df_b2c[df_b2c.columns[8]] = df_b2c[df_b2c.columns[8]].astype(str).apply(convert_datetime).astype('datetime64[ns]')
    df_b2c[df_b2c.columns[10]] = df_b2c[df_b2c.columns[10]].astype(str).apply(convert_datetime).astype('datetime64[ns]')
    df_b2c[df_b2c.columns[11]] = df_b2c[df_b2c.columns[11]].astype(str).apply(convert_datetime).astype('datetime64[ns]')
    df_b2c[df_b2c.columns[12]] = df_b2c[df_b2c.columns[12]].astype(str).apply(convert_datetime).astype('datetime64[ns]')

    with pd.ExcelWriter(output, engine='xlsxwriter') as writer:
        df_ref.to_excel(writer, sheet_name='1.1 Referral_Daily Input', index=False)
        df_partner.to_excel(writer, sheet_name='1.2 Referral_Partner List', index=False)
        df_chc.to_excel(writer, sheet_name='2. CHC_Daily Input', index=False)
        df_b2c.to_excel(writer, sheet_name='3. B2C_Daily Input', index=False)
```
